## 4. Listes

### Création de liste

```python
L = [3, 1, 4]
L
```

### Accéder à des éléments d'une liste

```python
print("Premier élément :", L[0])
print("Dernier élément :", L[-1])
print("Sous-liste [1:4] :", L[1:4])
```

### Modification d'un élément

```python
L[2] = 300
print("Après modification :", L)
```

### Ajout d'éléments

```python
L.append(60)               # ajoute à la fin
L.insert(2, 15)            # insère à l'indice 2
print("Après append et insert :", L)
```

### Suppression d'éléments

```python
del L[1]                   # supprime l’élément d’indice 1
val = L.pop()              # supprime le dernier élément et le renvoie
print("Après suppressions :", L, "| élément retiré :", val)
```

### Recherche et comptage

```python
L2 = [3, 1, 4, 1, 5, 9, 1]
print("Nombre de 1 dans L2 :", L2.count(1))
print("Indice du premier 4 :", L2.index(4))
```

### Extension de liste

```python
L3 = [100, 200]
L_comb = L + L3
print("Concaténation :", L_comb)
L.extend([300, 400])
print("Après extend :", L)
```

### Tri et inversion de liste

```python
L.sort()
print("Triée :", L)
L.reverse()
print("Inversée :", L)
```

### Compréhension de liste

```python
carres_pairs = [n*n for n in L]
print(carres_pairs)
```

### Copie vs référence

```python
L2 = L
L2.append(9)
print("Alias (L, L2)", L, L2)

L3 = L.copy()
L3.append(99)
print("Copie (L, L3)", L, L3)
```

```python
L
```

## 6. Dictionnaires

### Création de dictionnaire

```python
personne = {"nom": "Alice", "age": 30, "ville": "Paris"}
print("Dictionnaire :", personne)
```

### Accès à une valeur

```python
print("Nom :", personne["nom"])
```

### Ajout et modification

```python
personne["age"] = 31                   # modification
personne["profession"] = "ingénieure"  # ajout
print("Après ajout et modification :", personne)
```

### Suppression d'éléments

```python
del personne["ville"]
metier = personne.pop("profession")     # retire et renvoie la valeur
print("Après suppressions :", personne, "| métier retiré :", metier)
```

### Clés, valeurs

```python
print("Clés :", list(personne.keys()))
print("Valeurs :", list(personne.values()))
```

### Conversion liste $\rightarrow$ dictionnaire

```python
cles = ["a", "b", "c"]
valeurs = [1, 2, 3]
d = dict(zip(cles, valeurs))
print("Depuis deux listes :", d)
```

## 7. Conditions (if / elif / else)

### Condition simple

```python
# Condition simple
x = 10
if x > 0:
   print("x est positif")
```

### Condition avec else

```python
y = -5
if y >= 0:
    print("y est positif ou nul")
else:
    print("y est négatif")
```

### Condition avec elif

```python
note = 14
if note >= 16:
    mention = "Très bien"
elif note >= 12:
    mention = "Assez bien"
elif note >= 10:
    mention = "Passable"
else:
    mention = "Insuffisant"
print("Mention :", mention)
```

### Combinaison de conditions

```python
# Condition avec and
age = 25
permis = True
if age >= 18 and permis:
    print("Peut conduire")
else:
    print("Ne peut pas conduire")

# Condition avec or
jour = "dimanche"
if jour == "samedi" or jour == "dimanche":
    print("C’est le week-end")
else:
    print("C’est un jour de semaine")
```

### Conditions avec négation

```python
a = 3
if not a % 2 == 0:
    print("a est impair")
```

## 8. Boucles for

### Boucle for simple

```python
fruits = ["pomme", "banane", "cerise"]
for fruit in fruits:
    print("Fruit :", fruit)
```

### Boucle for avec `range`

```python
for i in range(5):  # de 0 à 4 (on commence avec 0 en Python !)
    print("i =", i)
```

### Boucle for avec pas

```python
for i in range(2, 11, 2):  # de 2 à 10 par pas de 2
    print("i pair =", i)
```

### Boucle for avec enumerate

```python
animaux = ["chat", "chien", "lapin"]
for i, nom in enumerate(animaux, start=1):
    print(f"Animal {i} :", nom)
```

### Boucle for sur un dictionnaire

```python
personne = {"nom": "Alice", "age": 30, "ville": "Lyon"}
for cle, valeur in personne.items():
    print(cle, ":", valeur)
```

### Boucle for sur plusieurs listes avec `zip`

```python
noms = ["Alice", "Bob", "Charlie"]
ages = [25, 30, 35]
for nom, age in zip(noms, ages):
    print(f"{nom} a {age} ans")
```

### Boucles imbriquées

```python
for i in range(1, 4):
    for j in range(1, 3):
        print(f"i={i}, j={j}")
```

```python
ma_liste = list(range(1, 4))
print(ma_liste)
len(ma_liste)
```

## 9. Fonctions

### Fonction simple

```python
def saluer():
    #Affiche un message de bienvenue
    print("Bonjour et bienvenue !")

saluer()
```

### Fonctions avec arguments

```python
def addition(a, b):
    """Renvoie la somme de deux nombres."""
    return a + b

resultat = addition(5, 3)
print("Résultat de l’addition :", resultat)
```

### Valeurs d'arguments par défaut

```python
def presentation(nom, age=25):
    """Affiche une présentation avec un âge par défaut."""
    print(f"Je m'appelle {nom} et j'ai {age} ans.")

presentation("Alice")
presentation("Bob", 32)
```

### Arguments nommés

```python
def afficher_coordonnees(x, y):
    print(f"Coordonnées : x={x}, y={y}")

afficher_coordonnees(y=20, x=10)
```

### Arguments variables `*args`

```python
def somme(*nombres):
    """Calcule la somme de plusieurs nombres."""
    return sum(nombres)

print("Somme :", somme(1, 2, 3, 4, 5))
```

### Arguments nommés variables `**kwargs`

```python
def afficher_infos(**infos):
    """Affiche les informations passées sous forme de mots-clés."""
    for cle, valeur in infos.items():
        print(cle, ":", valeur)

afficher_infos(nom="Alice", age=30, ville="Lyon")
```

### Variable globale vs variable locale

```python
x = 5
def afficher():
    #global x
    x = 10
    print("x dans la fonction :", x)

afficher()
print("x en dehors :", x)
```

### Fonction `lambda`

```python
carre = lambda n: n ** 2
print("Carré de 5 :", carre(5))
```
